```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.table import Table

# Dados da tabela
dados = {
    "Nome da Banda": ["B1", "B1", "B2", "B3", "B4", "B5", "B6", "B6", "B7", "B10, B11", "Pancromática", "QA_PIXEL"],
    "Descrição": [
        "Aerosol Costeiro", "Azul", "Verde", "Vermelho", "Infravermelho Próximo (NIR)",
        "Infravermelho de Ondas Curtas 1", "Infravermelho Térmico (TIR)",
        "Infravermelho de Ondas Curtas 2", "Infravermelho de Ondas Curtas 2",
        "Infravermelho Térmico (TIRS)", "Grayscale de Alta Resolução", "Qualidade do Pixel"
    ],
    "Landsat 4-5": ["—", "✔", "✔", "✔", "✔", "✔", "✔", "✔", "✔", "—", "—", "✔"],
    "Landsat 7":   ["—", "✔", "✔", "✔", "✔", "✔", "✔", "✔", "✔", "—", "✔", "✔"],
    "Landsat 8":   ["✔", "—", "✔", "✔", "✔", "✔", "—", "—", "✔", "✔", "✔", "✔"],
    "Comprimento de Onda (µm)": [
        "0,435 – 0,451", "0,45 – 0,52", "0,52 – 0,60", "0,63 – 0,69", "0,76 – 0,90",
        "1,55 – 1,75", "10,4 – 12,5", "2,08 – 2,35", "2,08 – 2,35",
        "10,60 – 12,51", "0,50 – 0,68", "—"
    ],
    "Resolução": ["30 m", "30 m", "30 m", "30 m", "30 m", "30 m", "120 m*", "30 m", "30 m", "100 m*", "15 m", "30 m"]
}

df = pd.DataFrame(dados)

# Criar figura
fig, ax = plt.subplots(figsize=(15, 20))
ax.axis('off')
tbl = Table(ax, bbox=[0, 0, 1, 1])

n_rows, n_cols = df.shape
widths = [0.08, 0.20, 0.05, 0.05, 0.05, 0.15, 0.07]
height = 0.15
font_size = 24

# Cabeçalhos
for i, col in enumerate(df.columns):
    tbl.add_cell(0, i, widths[i], height, text=col, loc='center', facecolor='#d9d9d9')  # cinza claro
    cell = tbl[(0, i)]
    cell.get_text().set_color('black')
    cell.get_text().set_weight('bold')
    cell.get_text().set_fontsize(font_size)

# Células
for row in range(n_rows):
    for col in range(n_cols):
        tbl.add_cell(row+1, col, widths[col], height, text=df.iat[row, col], loc='center', facecolor='white')
        cell = tbl[(row+1, col)]
        cell.get_text().set_fontsize(font_size)

ax.add_table(tbl)
plt.tight_layout()

# Salvar como imagem
output_path = "tabela_landsat.png"
plt.savefig(output_path, dpi=300, bbox_inches='tight')
output_path
```
